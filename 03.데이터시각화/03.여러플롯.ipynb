- Axes 객체와 subplot

```python
#### subplot 명령은 그리드(grid) 형태의 Axes 객체들을 생성하는데 Figure가 행렬(matrix)이고 Axes가 행렬의 원소라고 생각하면 된다. 예를 들어 위와 아래 두 개의 플롯이 있는 경우 행이 2 이고 열이 1인 2x1 행렬이다. 
#### subplot 명령은 세개의 인수를 가지는데 처음 두개의 원소가 전체 그리드 행렬의 모양을 지시하는 두 숫자이고 세번째 인수가 네 개 중 어느것인지를 의미하는 숫자이다. 
#### 따라서 위/아래 두개의 플롯을 하나의 Figure 안에 그리려면 다음처럼 명령을 실행해야 한다. 여기에서 숫자 인덱싱은 파이썬이 아닌 Matlab 관행을 따르기 때문에 첫번째 플롯을 가리키는 숫자가 0이 아니라 1임에 주의하라.
```

```python
subplot(2, 1, 1)  
# 여기에서 윗부분에 그릴 플롯 명령 실행

subplot(2, 1, 2)
# 여기에서 아랫부분에 그릴 플롯 명령 실행
```

```python
x1 = np.linspace(0., 5.)
x2 = np.linspace(0., 2.)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2,1,1)
plt.title('2 Subplots')
ax1.plot(x1,y1,'yo-')
ax1.set_ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
ax2.plot(x2,y2,'yo-')
ax2.set_ylabel('Undamped oscillation')

plt.xlabel('time(sec)')
ax1.grid();ax2.grid()
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 반복문 형태로 그래프 그릴 수 있도록 연습하기!
# 만약 그래프 8개라면 for i in range(8): ax = axes[i//4, i%4] 적용 
# ex) 그래프 4개일 때:
# 0 1 2 3
# (0,0) (0,1) (1,0) (1,1)

np.random.seed(0)

fig, axes = plt.subplots(2,2)
for i in range(4):
    ax = axes[i//2,i%2]            # [행 인덱스, 열 인덱스]
    ax.plot(np.random.rand(5))
    ax.set_title(f'axe {i+1}')
    ax.grid()

plt.tight_layout()
plt.show()
```
