```python
import folium
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
bic = pd.read_csv('top100.csv')
bic.head(3)
```

```python
bic.isnull().sum()
```

```python
bic['size']=''
```

```python
bic['size']=bic.대여건수/2000
```

```python
bic.head()
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)
for i in bic.index:
    folium.CircleMarker(
        location=[bic.위도[i], bic.경도[i]], 
        radius=int(bic['size'][i]),
        #tooltip=f"{bic.대여소명[i]}",
        color='#3186cc', fill_color='#3186cc'
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>따릉이 대여소명과 이용량</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
