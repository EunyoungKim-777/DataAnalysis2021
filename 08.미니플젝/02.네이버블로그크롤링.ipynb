```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
driver = webdriver.Chrome('../05.Crawling/chromedriver')
driver.get('https://blog.naver.com')
```

```python
search_box = driver.find_element_by_css_selector('input.textbox.ng-pristine.ng-untouched.ng-valid.ng-empty.ng-valid-maxlength')
search_box.send_keys('따릉이 타고')
search_box.send_keys(Keys.ENTER)
```

```python
# 페이지 지정
anchors = driver.find_elements_by_css_selector('a.item')
len_anchors = len(anchors)
len_anchors
```

```python
# 다음 페이지로 넘어가기 버튼
driver.find_elements_by_css_selector('a.button_next')[0].click()
```

```python
title_list=[]
j = 0
while j<50:

    for page in range(21,31):
        anchors = driver.find_elements_by_css_selector('a.item')
        anchors[page].click()
        print(page)
        time.sleep(1)

        for i in range(7):
            title = driver.find_elements_by_css_selector('.title_post')[i].text
            print(i,title)
            title_list.append(title)
            time.sleep(0.5)

    driver.find_elements_by_css_selector('a.button_next')[0].click()
    time.sleep(1)
    j += 1
```

```python
df = pd.DataFrame({"title":title_list})
df
```

```python
df.to_csv('title.txt', sep = '\t')
```
