```python
import pandas as pd
import folium
import numpy as numpy
import seaborn as sns
import matplotlib as mpl
```

```python
mpl.rc('font', family='NanumBarunpen')
mpl.rc('axes', unicode_minus=False)
```

```python
park = pd.read_csv('sppark.csv',encoding='EUC-KR')
park.head(1)
```

```python
newpark=park[['공원명','소재지지번주소','위도','경도','공원면적']]
newpark.head(1)
```

```python
newpark.isnull().sum()
```

```python
type(newpark.공원면적[0])
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)
for i in newpark.index:
    folium.CircleMarker(
        location=[newpark.위도[i], newpark.경도[i]], 
        radius=float( (newpark['공원면적'][i])**(1/2)*(0.1) ),
        tooltip=f"{newpark.공원명[i]}({newpark.공원면적[i]}㎡)",
        color='#3186cc', fill_color='#3186cc'
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>송파구 공원 위치와 크기</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
