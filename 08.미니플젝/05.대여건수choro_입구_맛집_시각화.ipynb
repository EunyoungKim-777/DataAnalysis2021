## 따릉이 지역구별 대여건수-choro
## 자전거도로입구, 송파구 맛집 표시

```python
import json
import folium
import warnings
#warnings.simplefilter(action='ignore',category=FutureWarning)
import numpy as np
import pandas as pd
```

```python
bi_data = pd.read_csv('sum.csv')
```

```python
bi_data.head()
```

```python
bi_data=bi_data.set_index('지역구')
```

```python
bi_data.head()
```

```python
geo_path = '../04.지도_공공API/skorea_municipalities_geo_simple.json'
geo_data = json.load(open(geo_path,encoding='utf-8'))
```

```python
ent = pd.read_csv('ent.csv')
```

```python
mydf=pd.read_csv('newmj_393.csv')
```

```python
ent.head()
```

```python
mydf.head()
```

```python
mydf=mydf.drop(['Unnamed: 0'],axis=1)
```

```python
mydf.head()
```

```python
#map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map = folium.Map(location=[37.5502, 126.982], zoom_start=11)

map.choropleth(
    geo_data = geo_data,
    data = bi_data.대여건수,
    columns = [bi_data.index, bi_data.대여건수],
    fill_color = 'PuRd',
    #fill_opacity = 0.4,
    key_on = 'feature.id'
)

for i in ent.index:
    folium.CircleMarker(
        location=[ent.위도[i], ent.경도[i]], 
        radius=10,
        tooltip=f"{ent.자전거도로입구명[i]}",
        color='black', fill_color='black'
    ).add_to(map)

for i in mydf.index:
    folium.Circle(
        location = [mydf.lat[i], mydf.lng[i]], 
        radius = 20,
        #radius=int(mydf['score'][i]),
        tooltip = mydf.title[i],
        color='blue', fill_color='blue'
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>맛집과 choro</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
